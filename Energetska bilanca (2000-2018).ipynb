```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open("Podatki/uvoz_izvoz.csv") as csvfile:
        reader = csv.DictReader(csvfile,delimiter=";")
        tabela = list(reader)

        leta = [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]        
        domaca_proizvodnja = []
        uvoz = []
        izvoz = []
        
        for i in range(1, len(tabela)):
            vrstica = list(tabela[i].values())
            del vrstica[0]
            domaca_proizvodnja.append(int(vrstica[0][0]))
            uvoz.append(int(vrstica[0][1]))
            izvoz.append(-int(vrstica[0][2]))
        
        plt.figure(figsize=(15,4))
        plt.plot(leta, domaca_proizvodnja, label="Domaca proizvodnja")
        plt.plot(leta, uvoz, label="Uvoz")
        plt.plot(leta, izvoz, label="Izvoz")
        plt.xlabel('Leto')
        plt.ylabel('1000 toe')
        plt.title("Energetska bilanca")
        plt.xticks(leta)
        ax = plt.gca()
        plt.legend(bbox_to_anchor=(1.0, 1.0), bbox_transform=ax.transAxes)
        plt.show()
```
