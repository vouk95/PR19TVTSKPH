```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# gospodinjski odjemalci, z ddv
with open("Podatki/cene_plin_energija_gosp_ddv.csv") as csvfile:
        reader = csv.DictReader(csvfile,delimiter=";")
        tabela = list(reader)

        leta = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
        plin = []
        energija = []
        
        for i in range(0, len(tabela)):
            if i == 1:
                plin = list(tabela[i].values())
                del plin[0]
                plin = [int(j) for j in plin[0]]
            if i == 2:
                energija = list(tabela[i].values())
                del energija[0]
                energija = [int(j) for j in energija[0]]
        
        plin_leta = []
        for i, k in zip(plin[1:-1:2], plin[2:-1:2]):
            plin_leta.append(float((i + k) / 2))
            
        energija_leta = []
        for i, k in zip(energija[1:-1:2], energija[2:-1:2]):
            energija_leta.append(float((i + k) / 2))
        
        plt.plot(leta, plin_leta, color='blue', label='Zemeljski plin')
        plt.plot(leta, energija_leta, color='orange', label='Energija')
        plt.xlabel('Leto')
        plt.ylabel('Delezi v EUR (%)')
        plt.title("Primerjava cen zemeljskega plina in elektrike za gospodinjske odjemalce (z DDV) v primerjavi z navadnim povprejem v EU")
        plt.legend()
        plt.show()
        
# negospodinjski odjemalci, brez ddv
with open("Podatki/cene_plin_energija_negosp_brezddv.csv") as csvfile:
        reader = csv.DictReader(csvfile,delimiter=";")
        tabela = list(reader)

        leta = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
        plin = []
        energija = []
        
        for i in range(0, len(tabela)):
            if i == 1:
                plin = list(tabela[i].values())
                del plin[0]
                plin = [int(j) for j in plin[0]]
            if i == 2:
                energija = list(tabela[i].values())
                del energija[0]
                energija = [int(j) for j in energija[0]]
        
        plin_leta = []
        for i, k in zip(plin[1:-1:2], plin[2:-1:2]):
            plin_leta.append(float((i + k) / 2))
            
        energija_leta = []
        for i, k in zip(energija[1:-1:2], energija[2:-1:2]):
            energija_leta.append(float((i + k) / 2))
        
        plt.plot(leta, plin_leta, color='blue', label='Zemeljski plin')
        plt.plot(leta, energija_leta, color='orange', label='Energija')
        plt.xlabel('Leto')
        plt.ylabel('Delezi v EUR (%)')
        plt.title("Primerjava cen zemeljskega plina in elektrike za negospodinjske odjemalce (brez DDV) v primerjavi z navadnim povprejem v EU")
        plt.legend()
        plt.show()
```
