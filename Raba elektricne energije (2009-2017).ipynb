```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open("Podatki/raba_el_energije.csv") as csvfile:
        reader = csv.DictReader(csvfile,delimiter=";")
        tabela = list(reader)

        leta = [2009,2010,2011,2012,2013,2014,2015,2016,2017]
        ogrevanje_prostorov = []
        hlajenje_prostorov = []
        ogrevanje_vode = []
        kuhanje = []
        razsvetljava = []
        hladilniki = []
        zmrz_skrinje = []
        pralni_stroji = []
        susilni_stroji = []
        pomivalni_stroji = []
        racunalniki = []
        televizije = []
        drugo = []
        
        for i in range(0, len(tabela)):
            if i == 1:
                ogrevanje_prostorov = list(tabela[i].values())
                del ogrevanje_prostorov[0]
                ogrevanje_prostorov = [int(j) for j in ogrevanje_prostorov[0]]
                
            if i == 2:
                hlajenje_prostorov = list(tabela[i].values())
                del hlajenje_prostorov[0]
                hlajenje_prostorov = [int(j) for j in hlajenje_prostorov[0]]
                
            if i == 3:
                ogrevanje_vode = list(tabela[i].values())
                del ogrevanje_vode[0]
                ogrevanje_vode = [int(j) for j in ogrevanje_vode[0]]
                
            if i == 4:
                kuhanje = list(tabela[i].values())
                del kuhanje[0]
                kuhanje = [int(j) for j in kuhanje[0]]
                
            if i == 5:
                razsvetljava = list(tabela[i].values())
                del razsvetljava[0]
                razsvetljava = [int(j) for j in razsvetljava[0]]
                
            if i == 6:
                hladilniki = list(tabela[i].values())
                del hladilniki[0]
                hladilniki = [int(j) for j in hladilniki[0]]
                
            if i == 7:
                zmrz_skrinje = list(tabela[i].values())
                del zmrz_skrinje[0]
                zmrz_skrinje = [int(j) for j in zmrz_skrinje[0]]
                
            if i == 8:
                pralni_stroji = list(tabela[i].values())
                del pralni_stroji[0]
                pralni_stroji = [int(j) for j in pralni_stroji[0]]
                
            if i == 9:
                susilni_stroji = list(tabela[i].values())
                del susilni_stroji[0]
                susilni_stroji = [int(j) for j in susilni_stroji[0]]
                
            if i == 10:
                pomivalni_stroji = list(tabela[i].values())
                del pomivalni_stroji[0]
                pomivalni_stroji = [int(j) for j in pomivalni_stroji[0]]
                
            if i == 11:
                racunalniki = list(tabela[i].values())
                del racunalniki[0]
                racunalniki = [int(j) for j in racunalniki[0]]
                
            if i == 12:
                televizije = list(tabela[i].values())
                del televizije[0]
                televizije = [int(j) for j in televizije[0]]
                
            if i == 13:
                drugo = list(tabela[i].values())
                del drugo[0]
                drugo = [int(j) for j in drugo[0]]
                
            temperatura = []
            for j, k in zip(ogrevanje_prostorov, hlajenje_prostorov):
                temperatura.append((i + k) / 2)
            
            stroji = []
            for j, k, l in zip(pralni_stroji, susilni_stroji, pomivalni_stroji):
                stroji.append((j + k + l) / 3)

                
        # 1. graf
        plt.plot(leta, temperatura, color='brown', label='Ogrevanje/hlajenje prostorov')
        plt.plot(leta, ogrevanje_vode, color='red', label='Ogrevanje vode')
        plt.plot(leta, kuhanje, color='orange', label='Kuhanje')
        plt.plot(leta, razsvetljava, color='gold', label='Razsvetljava')
        plt.plot(leta, hladilniki, color='olive', label='Hladilniki')
        plt.plot(leta, zmrz_skrinje, color='yellowgreen', label='Zamrzovalne skrinje')
        plt.plot(leta, stroji, color='y', label='Stroji (pralni, susilni, pomivalni)')
        plt.plot(leta, racunalniki, color='violet', label='Racunalniki')
        plt.plot(leta, televizije, color='m', label='Televizije')
        plt.plot(leta, drugo, color='c', label='Drugo')
        plt.xlabel('Leto')
        plt.ylabel('GWh')
        plt.title("Poraba elektricne energije")
        plt.ylim([0,1100])
        ax = plt.gca()
        plt.legend(bbox_to_anchor=(1.0, 1.0), bbox_transform=ax.transAxes)
        plt.show()
        
        # 2. graf
        barWidth = 0.08
        r1 = np.arange(len(leta))
        r2 = [x+0.01 + barWidth for x in r1]
        r3 = [x+0.01 + barWidth for x in r2]
        r4 = [x+0.01 + barWidth for x in r3]
        r5 = [x+0.01 + barWidth for x in r4]
        r6 = [x+0.01 + barWidth for x in r5]
        r7 = [x+0.01 + barWidth for x in r6]
        r8 = [x+0.01 + barWidth for x in r7]
        r9 = [x+0.01 + barWidth for x in r8]
        r10 = [x+0.01 + barWidth for x in r9]
        
        plt.figure(figsize=(15,4))
        plt.bar(r1, temperatura, color='brown', width=barWidth, edgecolor='white', label='Ogrevanje/hlajenje prostorov')
        plt.bar(r2, ogrevanje_vode, color='red', width=barWidth, edgecolor='white', label='Ogrevanje vode')
        plt.bar(r3, kuhanje, color='orange', width=barWidth, edgecolor='white', label='Kuhanje')
        plt.bar(r4, razsvetljava, color='gold', width=barWidth, edgecolor='white', label='Razsvetljava')
        plt.bar(r5, hladilniki, color='olive', width=barWidth, edgecolor='white', label='Hladilniki')
        plt.bar(r6, zmrz_skrinje, color='yellowgreen', width=barWidth, edgecolor='white', label='Zamrzovalne skrinje')
        plt.bar(r7, stroji, color='y', width=barWidth, edgecolor='white', label='Stroji (pralni, susilni, pomivalni)')
        plt.bar(r8, racunalniki, color='violet', width=barWidth, edgecolor='white', label='Racunalniki')
        plt.bar(r9, televizije, color='m', width=barWidth, edgecolor='white', label='Televizije')
        plt.bar(r10, drugo, color='c', width=barWidth, edgecolor='white', label='Drugo')

        plt.xlabel('Leto')
        plt.ylabel('GWh')
        plt.xticks([r + barWidth for r in range(len(leta))], leta)
        plt.title("Poraba elektricne energije")
        plt.ylim([0,1100])
        ax = plt.gca()
        plt.legend(bbox_to_anchor=(1.0, 1.0), bbox_transform=ax.transAxes)
        plt.show()
        
        # 3. in 4. graf
        L2009 = [temperatura[0], ogrevanje_vode[0], kuhanje[0], razsvetljava[0], hladilniki[0], zmrz_skrinje[0], 
                 stroji[0], racunalniki[0], televizije[0], drugo[0]]
        L2017 = [temperatura[-1], ogrevanje_vode[-1], kuhanje[-1], razsvetljava[-1], hladilniki[-1], zmrz_skrinje[-1], 
                 stroji[-1], racunalniki[-1], televizije[-1], drugo[-1]]
        labels = 'Ogrevanje/hlajenje prostorov', 'Ogrevanje vode', 'Kuhanje', 'Razsvetljava', 'Hladilniki', 'Zamrzovalne skrinje', 'Stroji (pralni, susilni, pomivalni)', 'Racunalniki', 'Televizije', 'Drugo'
        colors = ['brown', 'gold', 'orange', 'olive', 'yellowgreen', 'y', 'violet', 'm', 'c', 'dodgerblue']
        
        plt.pie(L2009, labels=labels, colors=colors, autopct='%1.1f%%', startangle=0)
        plt.title("Raba el. energije v letu 2009")
        plt.show()
        
        plt.pie(L2017, labels=labels, colors=colors, autopct='%1.1f%%', startangle=0)
        plt.title("Raba el. energije v letu 2017")
        plt.show()
        
        
        L2009vse = 0
        L2010vse = 0
        L2011vse = 0
        L2012vse = 0
        L2013vse = 0
        L2014vse = 0
        L2015vse = 0
        L2016vse = 0
        L2017vse = 0
        
        for i in range(0, 9):
            if i == 0:
                L2009vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 1:
                L2010vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 2:
                L2011vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 3:
                L2012vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 4:
                L2013vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 5:
                L2014vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 6:
                L2015vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 7:
                L2016vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
            if i == 8:
                L2017vse = temperatura[i] + ogrevanje_vode[i] + kuhanje[i] + razsvetljava[i] + hladilniki[i] + zmrz_skrinje[i] + stroji[i] + racunalniki[i] + televizije[i] + drugo[i]
        
        print("Poraba elektricne energije cez leta")
        i = 0
        print("===============================================================================================================================")
        print("| Leto | Temperatura | Voda | Kuhanje | Razsvetljava | Hladilniki | Zamrzovalniki | Stroji | Racunalniki | Televizije | Drugo |")
        print("===============================================================================================================================")
        print("| 2009 |           {} |   {} |       {} |           {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2009vse)*100), int((ogrevanje_vode[i]/L2009vse)*100), int((kuhanje[i]/L2009vse)*100), int((razsvetljava[i]/L2009vse)*100), int((hladilniki[i]/L2009vse)*100), int((zmrz_skrinje[i]/L2009vse)*100), 
                 int((stroji[i]/L2009vse)*100), int((racunalniki[i]/L2009vse)*100), int((televizije[i]/L2009vse)*100), int((drugo[i]/L2009vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2010 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2010vse)*100), int((ogrevanje_vode[i]/L2010vse)*100), int((kuhanje[i]/L2010vse)*100), int((razsvetljava[i]/L2010vse)*100), int((hladilniki[i]/L2010vse)*100), int((zmrz_skrinje[i]/L2010vse)*100), 
                 int((stroji[i]/L2010vse)*100), int((racunalniki[i]/L2010vse)*100), int((televizije[i]/L2010vse)*100), int((drugo[i]/L2010vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2011 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2011vse)*100), int((ogrevanje_vode[i]/L2011vse)*100), int((kuhanje[i]/L2011vse)*100), int((razsvetljava[i]/L2011vse)*100), int((hladilniki[i]/L2011vse)*100), int((zmrz_skrinje[i]/L2011vse)*100), 
                 int((stroji[i]/L2011vse)*100), int((racunalniki[i]/L2011vse)*100), int((televizije[i]/L2011vse)*100), int((drugo[i]/L2011vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2012 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2012vse)*100), int((ogrevanje_vode[i]/L2012vse)*100), int((kuhanje[i]/L2012vse)*100), int((razsvetljava[i]/L2012vse)*100), int((hladilniki[i]/L2012vse)*100), int((zmrz_skrinje[i]/L2012vse)*100), 
                 int((stroji[i]/L2012vse)*100), int((racunalniki[i]/L2012vse)*100), int((televizije[i]/L2012vse)*100), int((drugo[i]/L2012vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2013 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2013vse)*100), int((ogrevanje_vode[i]/L2013vse)*100), int((kuhanje[i]/L2013vse)*100), int((razsvetljava[i]/L2013vse)*100), int((hladilniki[i]/L2013vse)*100), int((zmrz_skrinje[i]/L2013vse)*100), 
                 int((stroji[i]/L2013vse)*100), int((racunalniki[i]/L2013vse)*100), int((televizije[i]/L2013vse)*100), int((drugo[i]/L2013vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2014 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2014vse)*100), int((ogrevanje_vode[i]/L2014vse)*100), int((kuhanje[i]/L2014vse)*100), int((razsvetljava[i]/L2014vse)*100), int((hladilniki[i]/L2014vse)*100), int((zmrz_skrinje[i]/L2014vse)*100), 
                 int((stroji[i]/L2014vse)*100), int((racunalniki[i]/L2014vse)*100), int((televizije[i]/L2014vse)*100), int((drugo[i]/L2014vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2015 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2015vse)*100), int((ogrevanje_vode[i]/L2015vse)*100), int((kuhanje[i]/L2015vse)*100), int((razsvetljava[i]/L2015vse)*100), int((hladilniki[i]/L2015vse)*100), int((zmrz_skrinje[i]/L2015vse)*100), 
                 int((stroji[i]/L2015vse)*100), int((racunalniki[i]/L2015vse)*100), int((televizije[i]/L2015vse)*100), int((drugo[i]/L2015vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2016 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2016vse)*100), int((ogrevanje_vode[i]/L2016vse)*100), int((kuhanje[i]/L2016vse)*100), int((razsvetljava[i]/L2016vse)*100), int((hladilniki[i]/L2016vse)*100), int((zmrz_skrinje[i]/L2016vse)*100), 
                 int((stroji[i]/L2016vse)*100), int((racunalniki[i]/L2016vse)*100), int((televizije[i]/L2016vse)*100), int((drugo[i]/L2016vse)*100)))
        i += 1
        print("|------+-------------+------+---------+--------------+------------+---------------+--------+-------------+------------+-------|")
        print("| 2017 |           {} |   {} |       {} |            {} |          {} |             {} |      {} |           {} |          {} |    {} |".format(int((temperatura[i]/L2017vse)*100), int((ogrevanje_vode[i]/L2017vse)*100), int((kuhanje[i]/L2017vse)*100), int((razsvetljava[i]/L2017vse)*100), int((hladilniki[i]/L2017vse)*100), int((zmrz_skrinje[i]/L2017vse)*100), 
                 int((stroji[i]/L2017vse)*100), int((racunalniki[i]/L2017vse)*100), int((televizije[i]/L2017vse)*100), int((drugo[i]/L2017vse)*100)))
        i += 1
        print("===============================================================================================================================")
        
        
```
