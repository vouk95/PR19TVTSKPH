# Poraba energije in goriv v gospodinjstvih letno

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open("Podatki/Poraba energije in goriv v gospodinjstvih letno.csv") as csvfile:
        reader = csv.DictReader(csvfile,delimiter=";")
        tabela = list(reader)
        leta   = list(tabela[0].keys())
        del leta[0]
        
        energetski_viri_skupaj = []
        ekstra_kurilno_olje = []
        zemeljski_plin = []
        lesna_goriva = []
        utekocinjeni_naftni_plin = []
        elektricna_energija = []
        premog = []
        daljinska_toplota = []
        soncna_energija = []
        toplota_iz_okolice = []
        for i in range(0, len(tabela)):
            if(i == 0):
                energetski_viri_skupaj = list(tabela[i].values())
                del energetski_viri_skupaj[0]
                energetski_viri_skupaj = [int(j) for j in energetski_viri_skupaj]
            if(i == 1):
                ekstra_kurilno_olje = list(tabela[i].values())
                del ekstra_kurilno_olje[0]
                ekstra_kurilno_olje = [int(j) for j in ekstra_kurilno_olje]
                
            if(i == 2):
                zemeljski_plin = list(tabela[i].values())
                del zemeljski_plin[0]
                zemeljski_plin = [int(j) for j in zemeljski_plin]
                
            if(i == 3):
                lesna_goriva = list(tabela[i].values())
                del lesna_goriva[0]
                lesna_goriva = [int(j) for j in lesna_goriva]
                
            if(i == 4):
                utekocinjeni_naftni_plin = list(tabela[i].values())
                del utekocinjeni_naftni_plin[0]
                utekocinjeni_naftni_plin = [int(j) for j in utekocinjeni_naftni_plin]
                
            if(i == 5):
                elektricna_energija = list(tabela[i].values())
                del elektricna_energija[0]
                elektricna_energija = [int(j) for j in elektricna_energija]
                
            if(i == 6):
                premog = list(tabela[i].values())
                del premog[0]
                premog = [int(j) for j in premog]
            
            if(i == 7):
                daljinska_toplota = list(tabela[i].values())
                del daljinska_toplota[0]
                daljinska_toplota = [int(j) for j in daljinska_toplota]
                
            if(i == 8):
                soncna_energija = list(tabela[i].values())
                del soncna_energija[0]
                soncna_energija = [int(j) for j in soncna_energija]
                
            if(i == 9):
                toplota_iz_okolice = list(tabela[i].values())
                del toplota_iz_okolice[0] 
                toplota_iz_okolice = [int(j) for j in toplota_iz_okolice]
        print(energetski_viri_skupaj)
        print(toplota_iz_okolice)
        
        import numpy as np
        import matplotlib.pyplot as plt
        plt.plot(leta, energetski_viri_skupaj)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("energetski_viri_skupaj")
        plt.ylim([0, 60000])
        plt.show()
        
        plt.plot(leta, ekstra_kurilno_olje)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("ekstra_kurilno_olje")
        plt.ylim([0, 12000])
        plt.show()
        
        plt.plot(leta, zemeljski_plin)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("zemeljski_plin")
        plt.ylim([0, 6000])
        plt.show()
        
        plt.plot(leta, lesna_goriva)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("lesna_goriva")
        plt.ylim([0, 25000])
        plt.show()
        
        plt.plot(leta, utekocinjeni_naftni_plin)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("utekocinjeni_naftni_plin")
        plt.ylim([0, 2000])
        plt.show()
        
        plt.plot(leta, elektricna_energija)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("elektricna_energija")
        plt.ylim([0, 13000])
        plt.show()
        
        plt.plot(leta, premog)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("premog")
        plt.show()
        
        plt.plot(leta, daljinska_toplota)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("daljinska_toplota")
        plt.ylim([0, 5000])
        plt.show()
        
        plt.plot(leta, soncna_energija)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("soncna_energija")
        plt.ylim([0, 500])
        plt.show()
        
        plt.plot(leta, toplota_iz_okolice)
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("Toplota iz okolice")
        plt.ylim([0, 2000])
        plt.show()
                
```
