```python
import csv
import numpy as np
import matplotlib.pyplot as plt

with open("Podatki/poraba_energije.csv") as csvfile:
        reader = csv.DictReader(csvfile,delimiter=";")
        tabela = list(reader)

        leta = [2009,2010,2011,2012,2013,2014,2015,2016,2017]
        kurilno_olje = []
        zemeljski_plin = []
        les_drva = []
        les_paleti = []
        les_ostalo = []
        naftni_plin = []
        el_energija = []
        premog = []
        daljinska_toplota = []
        soncna_energija = []
        topl_iz_okolice = []
        
        for i in range(0, len(tabela)):
            if i == 1:
                kurilno_olje = list(tabela[i].values())
                del kurilno_olje[0]
                kurilno_olje = [int(j) for j in kurilno_olje[0]]
                
            if i == 2:
                # 100 kubicnih metrov (1000 Sm3) = 353 ton
                zemeljski_plin = list(tabela[i].values())
                del zemeljski_plin[0]
                zemeljski_plin = [int(j)*353 for j in zemeljski_plin[0]]
                
            if i == 3:
                les_drva = list(tabela[i].values())
                del les_drva[0]
                les_drva = [int(j) for j in les_drva[0]]
                
            if i == 4:
                les_paleti = list(tabela[i].values())
                del les_paleti[0]
                les_paleti = [int(j) for j in les_paleti[0]]
                
            if i == 5:
                les_ostalo = list(tabela[i].values())
                del les_ostalo[0]
                les_ostalo = [int(j) for j in les_ostalo[0]]
                
            if i == 6:
                naftni_plin = list(tabela[i].values())
                del naftni_plin[0]
                naftni_plin = [int(j) for j in naftni_plin[0]]
                
            if i == 7:
                # 1 GigaWatt/hour (GWh) = 123 ton premoga (TCE - tons of coal equivalent)
                el_energija = list(tabela[i].values())
                del el_energija[0]
                el_energija = [int(j)*123 for j in el_energija[0]]
                
            if i == 8:
                premog = list(tabela[i].values())
                del premog[0]
                premog = [int(j) for j in premog[0]]
                
            if i == 9:
                # 1 TeraJoule = 239 ton
                daljinska_toplota = list(tabela[i].values())
                del daljinska_toplota[0]
                daljinska_toplota = [int(j)*239 for j in daljinska_toplota[0]]
                
            if i == 10:
                # 1 TeraJoule = 239 ton
                soncna_energija = list(tabela[i].values())
                del soncna_energija[0]
                soncna_energija = [int(j)*239 for j in soncna_energija[0]]
                
            if i == 11:
                # 1 TeraJoule = 239 ton
                topl_iz_okolice = list(tabela[i].values())
                del topl_iz_okolice[0]
                topl_iz_okolice = [int(j)*239 for j in topl_iz_okolice[0]]
        
        # 1. graf
        plt.plot(leta, kurilno_olje, color='brown', label='Ekstra lahko kurilno olje')
        plt.plot(leta, zemeljski_plin, color='red', label='Zemeljski plin')
        plt.plot(leta, les_drva, color='orange', label='Drva - polena')
        plt.plot(leta, les_paleti, color='gold', label='Lesni peleti')
        plt.plot(leta, les_ostalo, color='olive', label='Lesni sekanci, briketi in ostanki')
        plt.plot(leta, naftni_plin, color='yellowgreen', label='Utekocinjeni naftni plin')
        plt.plot(leta, el_energija, color='y', label='Elektricna energija')
        plt.plot(leta, premog, color='violet', label='Premog')
        plt.plot(leta, daljinska_toplota, color='m', label='Daljinska toplota')
        plt.plot(leta, soncna_energija, color='c', label='Soncna energija')
        plt.plot(leta, topl_iz_okolice, color='dodgerblue', label='Toplota iz okolice')
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.title("Primerjava virov")
        plt.ylim([0,1400000])
        ax = plt.gca()
        plt.legend(bbox_to_anchor=(1.0, 1.0), bbox_transform=ax.transAxes)
        plt.show()
        
        # 2. graf
        barWidth = 0.1
        r1 = np.arange(len(leta))
        r2 = [x + barWidth for x in r1]
        r3 = [x + barWidth for x in r2]
        r4 = [x + barWidth for x in r3]
        r5 = [x + barWidth for x in r4]
        r6 = [x + barWidth for x in r5]
        r7 = [x + barWidth for x in r6]
        r8 = [x + barWidth for x in r7]
        r9 = [x + barWidth for x in r8]
        r10 = [x + barWidth for x in r9]
        r11 = [x + barWidth for x in r10]
        
        plt.bar(r1, kurilno_olje, color='brown', width=barWidth, edgecolor='white', label='Ekstra lahko kurilno olje')
        plt.bar(r2, zemeljski_plin, color='red', width=barWidth, edgecolor='white', label='Zemeljski plin')
        plt.bar(r3, les_drva, color='orange', width=barWidth, edgecolor='white', label='Drva - polena')
        plt.bar(r4, les_paleti, color='gold', width=barWidth, edgecolor='white', label='Lesni peleti')
        plt.bar(r5, les_ostalo, color='olive', width=barWidth, edgecolor='white', label='Lesni sekanci, briketi in ostanki')
        plt.bar(r6, naftni_plin, color='yellowgreen', width=barWidth, edgecolor='white', label='Utekocinjeni naftni plin')
        plt.bar(r7, el_energija, color='y', width=barWidth, edgecolor='white', label='Elektricna energija')
        plt.bar(r8, premog, color='violet', width=barWidth, edgecolor='white', label='Premog')
        plt.bar(r9, daljinska_toplota, color='m', width=barWidth, edgecolor='white', label='Daljinska toplota')
        plt.bar(r10, soncna_energija, color='c', width=barWidth, edgecolor='white', label='Soncna energija')
        plt.bar(r11, topl_iz_okolice, color='dodgerblue', width=barWidth, edgecolor='white', label='Toplota iz okolice')
        
        plt.xlabel('Leto')
        plt.ylabel('Ton')
        plt.xticks([r + barWidth for r in range(len(leta))], leta)
        plt.title("Primerjava virov")
        plt.ylim([0,1400000])
        ax = plt.gca()
        plt.legend(bbox_to_anchor=(1.0, 1.0), bbox_transform=ax.transAxes)
        
        plt.show()
```
